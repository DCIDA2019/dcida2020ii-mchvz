```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
import cmath
from numpy import random
```

```python
n=10000
random.seed(42)
x=random.rand(n)*100
sigma=2
mu=70
x=np.float32(x)

plt.plot(x,'.')
nb=50
g0=np.histogram(x,nb)[0]
```

```python
phi=np.zeros(len(g0))
phi=np.float64(phi)
gh=np.histogram(x,nb)[1]
g1=np.zeros(len(g0))

a=1000*((1/(sigma*math.sqrt(2*np.pi))))

for i in range(1,len(gh)):
    g1[i-1]=(gh[i]+gh[i-1])/2
    
    
for i in range(len(g0)):
    phi[i]=math.exp(-( (g1[i]-mu)**2/(2*sigma**2) ))
    phi[i]=phi[i]*a*10**200 *g0[i]
    
plt.plot(g1,phi)
```

```python
# la variable his es la cantidad de datos repetidos normalizados
his= np.histogram(phi,nb)[0]/n
phi=np.zeros(len(his))
# para hacer la suma acumulada hay que igualar los primeros valores
phi[0]=his[0]

for i in np.arange(1,len(his)):
    phi[i]=phi[i-1]+his[i]
    
plt.plot(phi)
# Se pierde la normalizacion no se en que momento, creo que al aplicar la transformación a gaussiana.
```

# Binomial

```python
# n es el numero de datos, r es el numero de repeticiones por cada exerimento, p es la probabilidad
# nb es el numero de barras
n=30000
np.random.seed(42)
r , p = 100 , 0.4
s = np.random.binomial(r, p, n)
nb=20
plt.hist(s,nb);
```

```python
# la variable his es la cantidad de datos repetidos normalizados
his= np.histogram(s,nb)[0]/n
phi=np.zeros(len(his))
# para hacer la suma acumulada hay que igualar los primeros valores
phi[0]=his[0]

for i in np.arange(1,len(his)):
    phi[i]=phi[i-1]+his[i]
    
plt.plot(phi)
```
