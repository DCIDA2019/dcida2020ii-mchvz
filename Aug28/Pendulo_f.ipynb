# Usando numpy

```python
#Se define la función
def F(x,y):
    f=-k*np.sin(x)+y
    return f
```

Runge - Kutta

```python
#Se define la función Runge-Kutta
#name es el título de la gráfica
def RK(name):

    for i in range(1,tl):
    
#------calculos------------------
        xn=x[i-1]
        vn=F(xn,0)
    
        k1=dt*F(xn,vn)
        k2=dt*F(xn+dt/2,vn+k1/2)
        k3=dt*F(xn+dt/2,vn+k2/2)
        k4=dt*F(xn+dt,vn+k3)
    
#-------nuevos valores-----------
        v[i]=v[i-1]+(k1+2*k2+2*k3+k4)/6
        x[i]=x[i-1]+dt*v[i-1]
        
    plt.plot(t, x, label='Posición')
    plt.plot(t, v, label='Velocidad')
    plt.title(name)
    plt.grid()
    plt.legend()
    plt.xlabel('Tiempo')
    plt.show()
```
