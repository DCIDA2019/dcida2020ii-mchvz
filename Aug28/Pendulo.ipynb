```python
import math as math
import numpy as np
import matplotlib.pyplot as plt
```

Variables

```python
g=9.81 #gravedad
l=1  #longitud del pendulo
w=(g/l)**(1/2)

dt=0.1 #intervalos de tiempo
t = np.arange(0,6+dt,dt) #tiempo
#print('t =', t)

tl=t.size  #time lenght
X=np.zeros(tl)
i=0  
#print('tl=',tl)
#print('X=',X)
```

```python
A=1 
B=0
x0=A  #posici√≥n inicial
vx0=B  #velocidad inicial
#ax0=
```

```python
for n in t:
    x=A*np.cos(w*n)
    X[i]=x
    i=i+1
    
print('X =', X)
```
